````python
results = [
    ("Otieno",  43.2),
    ("Brian",   44.8),
    ("Amina",   45.1),
    ("Wanjiku", 46.7),
    ("Kamau",   48.3),
]

for position, (name, time) in enumerate(results, start=1):
    if position == 1:
        suffix = "st"
        warning = " - WINNER"
    elif position == 2:
        suffix = "nd"
        warning = ""
    elif position == 3:
        suffix = "rd"
        warning = ""
    else:
        suffix = "th"
        warning = ""

    print(f"{position}{suffix}: {name} - {time}s{warning}")
````

# Nested Loops
- nested loop is a loop inside another loop

-  The "inner loop" executes completely for every single iteration of the "outer loop"
- for every ONE run of the outer loop, the inner loop runs COMPLETELY from start to finish

````python
# number grid example
for row in range(3): # 0, 1, 2
  for col in range(4):
    print(col, end=" ")
  print()

"""
Explanation:
  Outer loop = rows
  Inner loop = columns
so:
  Row 1 -> print 0 1 2 3
  Row 2 -> print 0 1 2 3
  Row 3 -> print 0 1 2 3
"""
````

````python
# Mulptiplication table
for x in range(1,4): # 0, 1, 2
  for y in range(1,4):
    print(f"{x} x {y} = {x * y}")
````

````python
for row in range(4):
  for col in range(5):
    print("*", end=" ")
  print()
````

````python
for row in range(10):
  for x in range(row):
    print("*", end=" ")
  print()

  """
  Explanation:
    First round - print 1 star (range 1)
    second round - print 2 stars
    third round  - print 3 stars
  """
````

````python
for row in range(4):
  for col in range(5):
    print("#", end=" ")
  print()
````

````python
"""
(0,0) (0,1) (0,2)
(1,0) (1,1) (1,2)
"""

scores = [
    [78, 85, 90], # Amina
    [65, 72, 68], # Tom
    [91, 88, 95], # Njeri
]

# think of it as table
"""
          col 0 col 1   col 2
          maths English Science
row 0 -> 78     85      90
row 1 ->65      72      68
row 2 ->  91    88      95
"""
````

````python
# class results for multiple subjects

students = ["Amina", "Tom", "Njeri"]
subjects = ["Maths", "English", "Science"]
scores = [
    [78, 85, 90], # Amina
    [65, 72, 68], # Tom
    [91, 88, 95], # Njeri
]

for i, student in enumerate(students):
  print(f"------ {student} -------")
  for s, subject in enumerate(subjects):
    # print(f"scores[{i}][{s}] - {scores[i][s]} - {student}, {subject}")
    score = scores[i][s]
    print(f"  {subject}: {score}")
  print()




# ---- Amina -----
# Maths: 78
# English: 85
# Science: 90
````

- A wedding planner arranging chairs: 3 tables , each with 4 chairs.(The outer loop is tables, inner loop is chairs)

````python
chairs = ["A", "B", "C", "D"]
tables = range(3)

for x, table in enumerate(tables, start=1):
    print(f"Table: {x}")
    for chair in chairs:
        print(f"Chair placement: {chair}")
    print()

"""
Table: 1:
Chair 1
Chair 2
Chair 3
Chair 4
"""
````

- Multiplication grid , rows 1 -3 , and for each row , we write columns 1 to 3. (3X3 multiplication grid)


```
1x1=1 1x2=2 1x3=3
2x1=2 2x2=4 2x3=6
..................
```

````python
for num_1 in range(1,4):
    for num_2 in range(1,4):
        print(f"{num_1}x{num_2}={num_1 * num_2}", end=" ")
    print()
````

````python
# Multiplication table grid
for row in range(1,4):
    for col in range(1,4):
        product = row * col
        print(f"{row}x{col}={product}", end=" ")
    print()
````

````python
print("---Multiplication Grid---")
for row in range(1,4):
    for col in range(1,4):
        print(f"{row}x{col}={row*col}", end=" ")
    print()
````
