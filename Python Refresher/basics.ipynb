8). A company made sales of 850000, paid expenses of 320000, and paid tax of 18% on profit. Calculate profit before tax, tax amount, and profit after tax.


```python
sales = 850000
expenses = 320000
tax_rate = .18
profit_before_tax = sales - expenses
profit_before_tax

```

```python
tax_amount = profit_before_tax * tax_rate
tax_amount
```

```python
profit_after_tax = profit_before_tax - tax_amount
profit_after_tax
```

 An API provider gives 50,000 free requests. A company makes 350,000 requests. Each extra 1,000 requests costs $0.40. Calculate total cost in USD and KSh at an exchange rate of 129.

```python
free_requests = 50000
requests_made = 350000
cost_in_usd = .40
exchange_rate = 129
requests_to_pay = requests_made - free_requests
requests_to_pay

```

```python
batch_requests_per_1000 = requests_to_pay / 1000
batch_requests_per_1000
```

```python
cost_of_requests_in_usd = batch_requests_per_1000 * cost_in_usd
cost_of_requests_in_usd
```

```python

cost_of_requests_in_kes = cost_of_requests_in_usd * exchange_rate
cost_of_requests_in_kes
```

A freelancer earned $2500. The exchange rate is 129, and Upwork charges a 10% service fee. Calculate the amount received in KSh after the fee.

```python
income_usd = 2500
exchange_rate = 129
upwork_fee = .1

amount_recieved = (income_usd - (income_usd * upwork_fee)) * exchange_rate
amount_recieved
```
