```python
#Q15.  Water bill calculator for Nairobi Water. Units used: 0–6 m³ → Ksh 55/m³ (min charge Ksh 330), 7–20 m³ → Ksh 75/m³, 21–50 m³ → Ksh 105/m³, 51+ → Ksh 140/m³. Add 16% VAT to all bills. Show breakdown.  

vat = .16
units = 51


if units <= 6:
    price_per_unit = 55
    charge_bf_tax = price_per_unit * units
    charge_af_tax = charge_bf_tax + (charge_bf_tax * vat)
    print(f"units charge after tax {charge_af_tax}")
    if charge_af_tax < 300:
        charges_total = 300
    else:
        charges_total = charge_af_tax
elif units >= 7 and units < 20:
    price_per_unit = 75
    charge_bf_tax = price_per_unit * units
    charge_af_tax = charge_bf_tax + (charge_bf_tax * vat)
    charges_total = charge_af_tax
elif units >= 21 and units < 50:
    price_per_unit = 105
    charge_bf_tax = price_per_unit * units
    charge_af_tax = charge_bf_tax + (charge_bf_tax * vat)
    charges_total = charge_af_tax

elif units >= 51:
    price_per_unit = 140
    charge_bf_tax = price_per_unit * units
    charge_af_tax = charge_bf_tax + (charge_bf_tax * vat)
    charges_total = charge_af_tax


print(f"units charges {charges_total}")




```

```python
#  Body Mass Index (BMI) classifier. Ask for weight (kg) and height (m). Calculate BMI = weight / height². Classify: Underweight (<18.5), Normal (18.5–24.9), Overweight (25–29.9), Obese (30+). Print BMI and classification. 

weight = 12
height = 1.7

BMI = weight / (height * height)

if BMI < 18.5:
    print (f"You are underweight, BMI = {BMI}")
elif BMI >= 18.5 and BMI <= 24.9 :
    print (f"You are Normal, BMI = {BMI}")
elif BMI >= 25 and BMI <= 29.9 :
    print (f"You are Overweight, BMI = {BMI}")
else:
    print (f"You are Obeese, BMI = {BMI}")

```

```python
# matatu has 10 seats. Conductor doesnt know how many passengers will come , but he knows: WHILE there are empty seats, Keep loading. When full the matatu leaves
#- inside loop - seats filled update, remaining seats outside loop - print (Matatu full! Let's go)


seats_filled = 0
empty_seats = 10



while seats_filled < 11:
    print('loading')
    seats_filled += 1
    print(seats_filled)
    qs = input("seats remaining: ")
    print(qs)
    empty_seats -= 1

print("fully loaded")


```

```python
# balance = 5000, withdrawal = 1000, outside loop - insufficient funds for another withdrawal

balance = 5000
withdrawal = 1000

while balance >= 1000:
    balance -= withdrawal
    print(f"new balance: {balance}")
print(f"Insufficient funds {balance}")

```
