```python
#Q15.  Water bill calculator for Nairobi Water. Units used: 0–6 m³ → Ksh 55/m³ (min charge Ksh 330), 7–20 m³ → Ksh 75/m³, 21–50 m³ → Ksh 105/m³, 51+ → Ksh 140/m³. Add 16% VAT to all bills. Show breakdown.  

vat = .16
units = 51


if units <= 6:
    price_per_unit = 55
    charge_bf_tax = price_per_unit * units
    charge_af_tax = charge_bf_tax + (charge_bf_tax * vat)
    print(f"units charge after tax {charge_af_tax}")
    if charge_af_tax < 300:
        charges_total = 300
    else:
        charges_total = charge_af_tax
elif units >= 7 and units < 20:
    price_per_unit = 75
    charge_bf_tax = price_per_unit * units
    charge_af_tax = charge_bf_tax + (charge_bf_tax * vat)
    charges_total = charge_af_tax
elif units >= 21 and units < 50:
    price_per_unit = 105
    charge_bf_tax = price_per_unit * units
    charge_af_tax = charge_bf_tax + (charge_bf_tax * vat)
    charges_total = charge_af_tax

elif units >= 51:
    price_per_unit = 140
    charge_bf_tax = price_per_unit * units
    charge_af_tax = charge_bf_tax + (charge_bf_tax * vat)
    charges_total = charge_af_tax


print(f"units charges {charges_total}")




```

```python
# matatu has 10 seats. Conductor doesnt know how many passengers will come , but he knows: WHILE there are empty seats, Keep loading. When full the matatu leaves
#- inside loop - seats filled update, remaining seats outside loop - print (Matatu full! Let's go)


seats_filled = 0
empty_seats = 10



while seats_filled < 11:
    print('loading')
    seats_filled += 1
    print(seats_filled)
    qs = input("seats remaining: ")
    print(qs)
    empty_seats -= 1

print("fully loaded")


```

```python
#Write a password checker: keep asking for the password until the user gets it right. The correct password is 'kenya2024'. Count and display how many attempts it took.

password = 'kenya2024'
attempts = 0

while True:
    password_attempt = input("Please enter your password: ")
    while password_attempt != password and attempts < 3:
        attempts += 1
        print(f'try again, Total attempts: {attempts}')
        password_attempt = input("Please enter your password: ")
        
    print("Correct password")
    break


```

You have a list of temperatures recorded each hour. Print each temperature with its hour number. Also highlight any temperature above 35°C as a warning.   
temperatures = [22, 25, 29, 33, 37, 38, 36, 31, 27, 24].  

```python
temperatures = [22, 25, 29, 33, 37, 38, 36, 31, 27, 24]

for hour, temp in enumerate (temperatures, start=1):
    if temp > 35:
        print(f" Hour {hour} : Temp {temp} - High temp")
    else:
        print(f" Hour {hour} : Temp {temp}")
```

```python
temperatures = [22, 25, 29, 33, 37, 38, 36, 31, 27, 24]
hour = 1

for temp in temperatures:
    if temp > 35:
        print(f"Hour: {hour} - Temp: {temp} - High temp warning")
        hour += 1
    else:
        print(f"Hour: {hour} - Temp: {temp}")
        hour += 1
```

```python
# CHALLENGE: You have a list of product names and their stock levels. Print a numbered inventory. Skip any product with stock = 0. Stop the entire list when you hit a product called 'DISCONTINUED'.

inventory = [
    ("Unga 2kg",    50),
    ("Sugar 1kg",   0),
    ("Rice 2kg",    30),
    ("DISCONTINUED",0),
    ("Bread",       20),
    ("Cooking Oil", 15),
]

for number, (name, stock) in enumerate(inventory, start=1):
    if name == 'DISCONTINUED':
        break
    if stock == 0 :
        number -= 1
        continue
    
    print(f"{number} - Name: {name} - Stock: {stock}")
        
    


```

```python
# You have a list of runner names and their finishing times (in seconds). Print the results as a race leaderboard — 1st place, 2nd place, etc. Highlight the winner.
# Already sorted fastest first
# OUTPUT: 4th: Wanjiku - 46.7s

results = [
    ("Otieno",  43.2),
    ("Brian",   44.8),
    ("Amina",   45.1),
    ("Wanjiku", 46.7),
    ("Kamau",   48.3),
]

position_tag = ["1st", "2nd", "3rd ", "4th", "5th"]

for position, (name, time) in enumerate(results):
    tag = position_tag[position]
    if position == 0:
        print(f"{tag} - Name: {name} - Time: {time} - Winner")
    else:
        print(f"{tag} - Name: {name} - Time: {time}")
```
