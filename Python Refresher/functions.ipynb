### Python Functions. 
A function is a block of code which only runs when it is called.  

A function can return data as a result.   

A function helps avoiding code repetition.  


#### Function Names.  
Function names follow the same rules as variable names in Python:   

A function name must start with a letter or underscore.  
A function name can only contain letters, numbers, and underscores.  
Function names are case-sensitive (myFunction and myfunction are different).  

```python
def hello_world():
    print("hello world")
```

```python
def farenheit_to_celsius(temp):
    return (temp - 32) * 5 / 9

farenheit_to_celsius(100)
```

```python
#mpesa charge calc

def mpesa_charges(amount):
    if amount <= 100:
        charge = 0
    elif amount <= 500:
        charge = 11
    elif amount <= 1000:
        charge = 27
    else:
        charge = 75
    return charge


print("Charge for 90:" , mpesa_charges(90))
print("Charge for 110:" , mpesa_charges(110))
print("Charge for 900:" , mpesa_charges(900))
print("Charge for 9000:" , mpesa_charges(9000))
```

```python
# Student report - 

def student_report(name, maths, english, science):
    total = maths + english + science
    average = total / 3

    if average >= 80:
        grade = "A"
    elif average >= 70:
        grade = "B"
    elif average >= 60:
        grade = "C"
    elif average >= 80:
        grade = "D"
    else:
        grade = "D"

    print(f"-------- {name} -----------")
    print(f"   Maths: {maths} English: {english} Science: {science}")
    print(f"   Total: {total}/300 . Average : {average:.1f}")
    print(f"   Grade: {grade}")

student_report("amina", 78, 90, 85)
student_report("Brian", 90, 90, 85)

```
